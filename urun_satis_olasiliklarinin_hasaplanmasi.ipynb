# PROBLEM

## Bir yatırım öncesinde gelecek ay ile ilgili satışların belirli değerlerde gerçekleşmesi olasılıkları belirlenmek isteniyor.

* Dağılımın **normal** olduğu biliniyor
* Aylık ortalama satış sayısı 80K, standart sapması 5K
* 90K'dan fazla satış yapma olasılığı nedir?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

**90K'dan fazla satış yapma olasılığı ve grafiği**

```python
1-norm.cdf(90,80,5)
```

```python
# grafik çizimi
x=np.linspace(80-4*5,80+4*5,1000)
y=norm.pdf(x,80,5)
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Normal Dağılım')
plt.fill_between(x, y, where=(x >= 90), color='red', alpha=0.5)
plt.axvline(x=90, color='r', linestyle='--')
plt.title('Aylık Satış Sayısının Normal Dağılımı')
plt.xlabel('Aylık Satış (K)')
plt.ylabel('Olasılık Yoğunluğu')
plt.legend()
plt.show()
```
